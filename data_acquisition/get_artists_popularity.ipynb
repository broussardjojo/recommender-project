```python
import spotipy
import api_setup
import pathlib
import json
```

```python
env_vars = api_setup.parse_api_kvs(pathlib.Path.cwd().parent / "api-keys")
auth_manager = spotipy.SpotifyClientCredentials(env_vars['client_id'], env_vars['client_secret'])
sp = spotipy.Spotify(client_credentials_manager=auth_manager, backoff_factor=2)
```

```python
def split_into_sublists(input_list, chunk_size=50):
    upto = 0
    output_lists = []
    while True:
        if upto + chunk_size >= len(input_list):
            output_lists.append(input_list[upto:-1])
            break
        else:
            output_lists.append(input_list[upto:upto+chunk_size])
            upto += chunk_size
    return output_lists
```

```python
with open("./data/artists/artist_uris.txt", "r") as artist_data:
    artist_uris = [uri.strip() for uri in artist_data.readlines()]

artist_uris_chunked = split_into_sublists(artist_uris, 50)

artist_uri_to_popularity = {}
for idx, artist_uri_chunk in enumerate(artist_uris_chunked):
    artists_info = sp.artists(artist_uri_chunk)
    print(f"Completed chunk {idx + 1} / {len(artist_uris_chunked)}")
    for artist in artists_info['artists']:
        artist_uri_to_popularity[artist['id']] = artist['popularity']
```

```python
with open('./data/artists/artist_uris_to_popularity', "w") as f:
    json.dump(artist_uri_to_popularity, f)
```
