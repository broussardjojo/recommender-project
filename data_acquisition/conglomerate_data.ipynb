```python
import pandas as pd
import json
import ijson
```

```python
track_features = []
track_features_objects = ijson.items(open('./data/tracks/track_features.txt'), prefix="", multiple_values=True, use_float=True)

track_features = []
for sublist in track_features_objects:
	for item in sublist:
		if item:
			track_features.append(item)
```

```python
track_features
```

```python
track_features_df = pd.DataFrame.from_records(track_features)
track_features_df
```

```python
# I forgot to save the artist URIs for each song! >:(
with open("./data/tracks/track_to_artist.txt") as f:
	track_to_artist = json.load(f)
track_to_artist_df = pd.DataFrame(track_to_artist, index=[0]).transpose()
track_features_df = track_features_df.join(track_to_artist_df, on="id")
track_features_df = track_features_df.rename(columns={0: 'artist_uri'})
track_features_df.head(10)
```

```python
# Joining artist uris to their popularity
with open("./data/artists/artist_uris_to_popularity") as f:
	artist_to_popularity = json.load(f)
artist_popularity_df = pd.DataFrame(artist_to_popularity, index=[0]).transpose()
track_features_df = track_features_df.join(artist_popularity_df, on='artist_uri')
track_features_df = track_features_df.rename(columns={0: 'artist_popularity'})
track_features_df = track_features_df.drop(columns=['type', 'uri', 'track_href', 'analysis_url', 'artist_uri'])
track_features_df.head(10)
```

```python
track_features_df.to_csv("./data/final/scraped_dataset.csv")
```
